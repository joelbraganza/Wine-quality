# Red Wine quality

Let's import the necessary libraries

```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Red wine datatset is small, approx 1500 observations with 12 variables.
Let's have a look at the data.

```python
red = pd.read_csv(r'D:\joel\UDEMY -\datasets and codes\Notebooks\.ipynb_checkpoints\Wine Quality dataset\winequality-red.csv')
print (red.shape)
red.head(6)
```

Just like White wine dataset, this too has 6 unique quality classes, let's look at the distribution, both in ration & graphically

```python
print (red['quality'].value_counts())
red['quality'].value_counts().plot(kind='bar')
```

The distribution isn't equal for each quality. Just like in White wine dataset, we will categorize the quality into binary: 1 good, 0 bad

```python
red['quality'] = red['quality'].replace([3,4,5],0.0)
red['quality'] = red['quality'].replace([6,7,8],1.0)
print (red['quality'].value_counts())
red['quality'].value_counts().plot(kind='bar')
```

After categorizing the quality into binary, the dataset seems somewhat balanced w.r.t quality.

If outliers are present, let's remove them.

### Outlier removal

```python
# let's try and remove some outliers from both datasets
for i in red.columns[:-1]:
    q25,q75 = np.percentile(red[i],[25,75])
    iqr = q75-q25
    maximum = q75+(iqr*1.5)
    minimum = q25-(iqr*1.5)
    red.loc[red[i]>maximum,i]=np.NaN
    red.loc[red[i]<minimum,i]=np.NaN
```

Check what is the percentage of missing data after Outlier removal

```python
print ((red.isnull().sum()/red.shape[0])*100) # missing values after outlier removal in red
```

Outliers removed, let's impute them

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
red_array = np.array(red)
red = pd.DataFrame(data = imputer.fit_transform(red_array),columns = red.columns)
red.isnull().sum()
```

Let's also have a look at the distribution of the data after Outlier removal

### Data Visualization

```python
for col in red.columns[:-1]:
    sns.distplot(red[col],norm_hist=True,color='blue',kde_kws={'label':'KDE'}) # histogram with the kernel-distribution line.
    plt.axvline(red[col].mean(), color = 'r', linestyle = 'dashed', linewidth = 2) # plotting mean line
    plt.axvline(red[col].median(),color='b',linewidth=2) # plotting median line
    plt.show()
```

We can see that the data is normally distributed with the mean & median lines being closer and some cases coinciding. Could be because of Outlier removal.

Let's check which features show a relationship with the quality of the Red wine

Feature relationship

### ANOVA test

```python
from scipy import stats
not_related=[]
related=[]

for col in red.columns[:-1]:
    fStat, pVal = stats.f_oneway(red[col], red['quality'])
    #print(col,' P-value: ',pVal,' f-statistic: ',fStat)
    if pVal>0.05:
        not_related.append(col)
    else:
        related.append(col)
        
print ('not related: ',not_related)
print ('related: ',related)
```

### Extra trees classifier

##### let's perform two more feature selection techniques
##### Let's perform a feature importance test
##### this test gives you a score for each feature of your data, the higher score shows more importance/relevance 
##### of that feature towards your target variable
##### it uses a tree-based classifiers approach

```python
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.ensemble import ExtraTreesClassifier

# on the red wine dataset
redX = red.iloc[:,:-1]
redy = red['quality']
redImportances = ExtraTreesClassifier().fit(redX,redy)
#print (redImportances.feature_importances_)

# Let's look at the scores visually

print ('Red wine- feature importance')
redValues = pd.Series(redImportances.feature_importances_, index = red.columns[:-1])
redValues.nlargest(red.shape[1]-1).plot(kind='bar')
plt.figsize=(10,15)
plt.show()
```

### Logistic regression - feature importance

```python
# let's perform both tests on red wine first
# source: https://towardsdatascience.com/feature-selection-using-regularisation-a3678b71e499
from sklearn.linear_model import Lasso, LogisticRegression # Lasso is L1 regularization
from sklearn.feature_selection import SelectFromModel

# fitting the feature selection model with the Logistic regression data
X = red.iloc[:,:-1]
sel_ = SelectFromModel(LogisticRegression(C=1, penalty='l2'))
sel_.fit(X, red['quality'])


# Visualising features that were kept by the lasso regularisation
print (sel_.get_support())

#  So True is for the features that lasso thought is important (non-zero features) 
# while False is for the features whose weights were shrinked to zero and are not important according to Lasso.


selected_feat = X.columns[(sel_.get_support())] # gives the remaining features

print ('important: ',selected_feat)
print ('unimportant: ',[i for i in red.columns[:-1] if i not in selected_feat])
```

### Random Forest feature selection

```python
from sklearn.ensemble import RandomForestClassifier
sel_ = SelectFromModel(RandomForestClassifier())

sel_.fit(X, red['quality'])

# Visualising features that were kept by regularisation
print (sel_.get_support())
selected_feat = X.columns[(sel_.get_support())] # gives the remaining features
print ('Random-forest features selected: ')
print (selected_feat)
print ('Random forest\'s not important features: ')
print ([c for c in red.columns[:-1] if c not in selected_feat])
```

'fixed acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'density' are the set of features concluded as unimportant in both Random forest feature selection & Logistic regression feature importance, 
Extra trees classifier & ANOVA test gives a very contrasting result.

```python
remove =['fixed acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'density'] 
red2 = red[[i for i in red.columns if i not in remove]] # a new red wine dataframe without these variables
print (red2.shape)
red2.head()
```

### Feature Normalization

```python
for c in red2.columns:
    red2[c] = (red2[c]-min(red2[c]))/(max(red2[c])-min(red2[c]))
    print (c,': ',min(red2[c]),max(red2[c]))
    
red2.head()    
```

Let's split the data into train & test.

```python
from sklearn.model_selection import train_test_split
from collections import Counter

X = red2.iloc[:,:-1]
y = red2['quality']

# Let's split the data into train & test before oversampling the training data
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
print ('Training shape: ',X_train.shape,y_train.shape)
print ('Testing shape: ',X_test.shape,y_test.shape)
```

Let's define an error-metric function to judge the model's classifications

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

def error_metric(y_true,y_preds):
    CM = pd.crosstab(y_true,y_preds) # the confusion matrix
    TN = CM.iloc[0,0]
    FN = CM.iloc[1,0]
    TP = CM.iloc[1,1]
    FP = CM.iloc[0,1]
    print (classification_report(y_true, y_preds))
    print (confusion_matrix(y_true,y_preds))
    print ('accuracy score: ',accuracy_score(y_true, y_preds)*100,'%') #check accuracy of model
    print ('False-Negative-Rate: ',(FN*100)/(FN+TP),'%') #False Negative rate 
    print ('true-positive-rate: ',((TP)*100)/(TP+FN),'%')
```

Let's also define a function to plot the AUC-ROC curve

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
from matplotlib import pyplot as plt
def plotRocAuc(fpr, tpr, thresholds,score):
       
    # Plot the "dumb model" line
    plt.plot([0, 1], [0, 1], linestyle='--')
    
    # Plot the model line
    plt.plot(fpr, tpr, marker='.')
    plt.text(0.75, 0.25, "AUC: " + str(round(score,2)))
    
    # show the plot
    plt.show()
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression().fit(X_train, y_train) 
LR_preds = LR_model.predict_proba(X_test)
print (error_metric(y_test,np.round((LR_preds)[:,1])))
LR_fpr,LR_tpr,LR_thresholds = roc_curve(y_test,LR_preds[:,1])
LR_score = roc_auc_score(y_test, LR_preds[:,1])
plotRocAuc(LR_fpr, LR_tpr, LR_thresholds,LR_score)
```

### Neural Network

For neural network, I will try to balance out little bit of the data which was less in quality 0.

```python
from imblearn.over_sampling import RandomOverSampler
os=RandomOverSampler(1)
X_train_os,y_train_os = os.fit_sample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
print ('New training data shape: ',X_train_os.shape,y_train_os.shape)
```

I will also oversample the data manually to provide more set of observations for training

```python
from sklearn.utils import shuffle
X1 = np.array(X_train_os[y_train_os==1])
X0 = np.array(X_train_os[y_train_os==0])
X = np.vstack([X0, X0, X0, X1, X1, X1])
Y = np.concatenate(([0]*3*len(X0), [1]*3*len(X1)))
X,Y = shuffle(X,Y)
X.shape,Y.shape
```

let's apply deep-learning using Keras

```python
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='sigmoid', input_shape=(X.shape[1],)))
    model.add(layers.Dense(64, activation='sigmoid'))
    model.add(layers.Dense(1, activation='sigmoid')) # because of multi-class labels
    # defining the optimizer and loss and accuracy parameters
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    return model

# testing how the individual model with sampLe parameters scores on test data
X_train_os2, x_val, y_train_os2, y_val = train_test_split(X,Y,train_size=0.7)
print (X_train_os2.shape, y_train_os2.shape)
print (x_val.shape,y_val.shape)
model = build_model()
history = model.fit(X_train_os2, y_train_os2, batch_size=64, epochs=5, validation_data=(x_val, y_val))

DL_preds = model.predict_proba(X_test)
print (DL_preds.shape)
test_results = model.evaluate(X_test,y_test)
for i in range(len(test_results)):
    print (model.metrics_names[i],': ',test_results[i])

DL_preds2 = [round(i[0]) for i in DL_preds]


import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

DL_fpr, DL_tpr, DL_thresholds = roc_curve(y_test, DL_preds)
DL_score = roc_auc_score(y_test, DL_preds)
error_metric(y_test,np.array(DL_preds2))
plotRocAuc(DL_fpr, DL_tpr, DL_thresholds, DL_score)
```
