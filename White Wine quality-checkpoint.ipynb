```python
white['quality'] = white['quality'].replace([3, 4, 5, 6], 0.0) # class 3 to 5 are bad quality
white['quality'] = white['quality'].replace([ 7, 8, 9], 1.0)          # class 7/8 are good quality
```

We can try to balance this dataset after, first we need to do some outlier removal with feature relationship w.r.t quality.

### Outlier removal

```python
for i in white.columns[:-1]:
    q25,q75 = np.percentile(white[i],[25,75])
    iqr = q75-q25
    maximum = q75+(iqr*1.5)
    minimum = q25-(iqr*1.5)
    white.loc[white[i]>maximum,i]=np.NaN
    white.loc[white[i]<minimum,i]=np.NaN
    

print ((white.isnull().sum()/white.shape[0])*100) # missing values after outlier removal in white dataset
```

Let's impute the missing values using KNN algorithm

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
white_array = np.array(white)
white = pd.DataFrame(data = imputer.fit_transform(white_array),columns = white.columns)
white.head()
```

Let's define an error-metric function to judge the model's classifications

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

def error_metric(y_true,y_preds):
    CM = pd.crosstab(y_true,y_preds) # the confusion matrix
    TN = CM.iloc[0,0]
    FN = CM.iloc[1,0]
    TP = CM.iloc[1,1]
    FP = CM.iloc[0,1]
    print (classification_report(y_true, y_preds))
    print (confusion_matrix(y_true,y_preds))
    print ('accuracy score: ',accuracy_score(y_true, y_preds)*100,'%') #check accuracy of model
    print ('False-Negative-Rate: ',(FN*100)/(FN+TP),'%') #False Negative rate 
    print ('true-positive-rate: ',((TP)*100)/(TP+FN),'%')
```

Let's also define a function to plot the AUC-ROC curve 

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
from matplotlib import pyplot as plt
def plotRocAuc(fpr, tpr, thresholds,score):
       
    # Plot the "dumb model" line
    plt.plot([0, 1], [0, 1], linestyle='--')
    
    # Plot the model line
    plt.plot(fpr, tpr, marker='.')
    plt.text(0.75, 0.25, "AUC: " + str(round(score,2)))
    
    # show the plot
    plt.show()
```
